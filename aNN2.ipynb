```python
import numpy as np

def and_not(x1, x2):
    W = np.array([1, -1])  
    X = np.array([x1, x2])  
    T = 1                  
    y = np.dot(W, X)       
    return 1 if y >= T else 0

print("X1 X2 | Y")
print("-------------")
print(f"0  0 | {and_not(0,0)}")
print(f"0  1 | {and_not(0,1)}")
print(f"1  0 | {and_not(1,0)}")
print(f"1  1 | {and_not(1,1)}")





```
